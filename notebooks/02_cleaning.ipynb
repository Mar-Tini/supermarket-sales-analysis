```python
import matplotlib.pyplot as plt

colonnes_numeriques = ventes_clean.select_dtypes(include=['int64', 'float64']).columns.drop(['ItemCode','CategoryCode', 'SaleType','Discount'], errors='ignore')

ventes_clean[colonnes_numeriques].plot(
    kind='box',
    subplots=True,
    layout=(1, len(colonnes_numeriques)),
    figsize=(4*len(colonnes_numeriques), 6),
    title='Boxplots des variables numériques',
    sharey=False, 
)
plt.tight_layout()
plt.show()
```

D'après le graphique, on observe de nombreuses valeurs extrêmes, en particulier pour les quantités. Cela rend difficile la décision de supprimer ou de remplacer ces valeurs aberrantes. 

On peut regrouper les données par quantité pour segmenter les ventes en fonction de cette majorité d'outliers

```python
import pandas as pd

colonnes_numeriques = ['QuantityKg', 'UnitPrice', 'WholesalePrice']
ventes_clean = ventes_clean[(ventes_clean[colonnes_numeriques] >= 0).all(axis=1)]

ventes_clean[colonnes_numeriques].describe().round(4)
```

On supprime toutes les valeurs inférieures ou égales à 0, car il n'est pas logique d'avoir des prix ou des quantités nuls ou négatifs.

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
import matplotlib.pyplot as plt
import numpy as np

# Extraire la variable et transformer en 2D
X = ventes_clean['QuantityKg'].values.reshape(-1, 1)

# Calculer les inerties pour k = 1 à 10
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Détecter le coude automatiquement
kneedle = KneeLocator(K, inertia, curve='convex', direction='decreasing')
optimal_k = kneedle.knee

print("Nombre optimal de clusters (k) :", optimal_k)

# Tracer la courbe du coude
plt.figure(figsize=(8,5))
plt.plot(K, inertia, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertia (somme des distances au carré)')
plt.title('Méthode du coude pour QuantityKg')
plt.xticks(K)
plt.grid(True)
plt.axvline(optimal_k, color='red', linestyle='--', label=f'k optimal = {optimal_k}')
plt.legend()
plt.show()
```

Après le clustering, nous constatons que k=3 segmente nos ventes en trois groupes distincts selon les quantités vendues.

```python

X = ventes_clean['QuantityKg'].values.reshape(-1, 1)


kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X)


ventes_clean['QuantityCluster'] = clusters
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

variables = ['QuantityKg', 'UnitPrice', 'WholesalePrice', 'LossRate']
clusters = ventes_clean['QuantityCluster'].unique()

# Créer une figure avec plusieurs sous-graphiques
fig, axes = plt.subplots(1, len(variables), figsize=(20,5), sharey=False)

for i, var in enumerate(variables):
    sns.boxplot(x='QuantityCluster', y=var, data=ventes_clean, ax=axes[i])
    axes[i].set_title(var)
    axes[i].set_xlabel('Cluster')
    axes[i].grid(True, linestyle='--', alpha=0.5)

plt.suptitle('Distribution des variables par cluster', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
