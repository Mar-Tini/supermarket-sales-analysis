```python

import pandas as pd
df2 = LoadData("../data/raw/annex2.csv", "csv").data
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce')
df2_filtered = df2[df2['Date'].dt.year.isin([2020])]

dfs = [
    LoadData("../data/raw/annex1.csv", "csv", size=1000).data,
    df2_filtered, 
    LoadData("../data/raw/annex3.csv", "csv", size=1000).data,
    LoadData("../data/raw/annex4.csv", "csv", size=1000).data
]

manager = LoadData("../data/raw/annex1.csv", "csv", size=1000)

# Fusion
ventes = manager.fusionner(dfs, on="Item Code", how="inner")

ventes.head()
```

```python
ventes.to_csv("../data/raw/ventes.csv", index=False)
```
